```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.utils import load_img

import os
import seaborn as sns
from matplotlib import pyplot as plt

from fnmatch import fnmatch
```

```python
def count_files(directory, search_string):    
    count = 0
    for file in os.listdir(directory):
        if fnmatch(file, "*" + search_string + "*"):
            count += 1
    return count
```

```python
def crop_directory_subdirectory(path, n_horizontal, n_vertical):

    #This function takes a "root" directory and crops all the images in all the subdirectories, crops it in n_pieces, and then removes the ooriginal image

    for subdir, dirs, files in os.walk(path):
        for file in files:       
            
            f = os.path.join(subdir, file)
            im = tf.keras.utils.load_img(f)
            
            imgwidth, imgheight = im.size
            

            trozos_h = int(imgwidth/n_horizontal)
            trozos_v = int(imgheight/n_vertical)
            
            

            for i in range(n_horizontal):
                for j in range(n_vertical):
                    
                    print(f)

                    # print("i = ", i)
                    # print("j = ", j)
                    #print( trozos_v*j ,trozos_h*i ,trozos_v*j + trozos_v,trozos_h*i+ trozos_h)        
                        
                    trozo = im.crop((trozos_h*j ,trozos_v*i ,trozos_h*j + trozos_h,trozos_v*i+ trozos_v))    
                    

                    trozo.save(f + str(i) + "_" + str(j)+ ".jpg")


            os.remove(f)

            
```

```python
import os

def rename_files(folder_path):
    for root, _, files in os.walk(folder_path):
        for filename in files:
            # Get the full path of the file
            file_path = os.path.join(root, filename)

            # Check if the path represents a file (not a directory)
            if os.path.isfile(file_path):
                # Split the filename into base name and extension
                base_name, extension = os.path.splitext(filename)

                # Add an underscore four characters before the extension
                if len(base_name) > 4:
                    new_base_name = base_name[:-3] + '_' + base_name[-3:]
                else:
                    new_base_name = '_' + base_name

                # Construct the new filename
                new_filename = new_base_name + extension

                # Construct the new path with the renamed file
                new_file_path = os.path.join(root, new_filename)

                # Rename the file
                os.rename(file_path, new_file_path)

# Specify the path to the main folder containing the subfolders and files
folder_path = 'new_private_dataset'

# Call the function to rename the files
rename_files(folder_path)
```
