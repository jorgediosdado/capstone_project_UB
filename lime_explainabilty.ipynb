```python
import gc
gc.collect()
```

```python
input_folder = 'to_explain_images'
output_folder = 'LIME_explained_images'



# Get the list of image files in the input folder
image_files = os.listdir(input_folder)

# Loop through each image file
for image_file in image_files:

    gc.collect()
    # Construct the input and output file paths
    input_image_path = os.path.join(input_folder, image_file)
    output_image_path = os.path.join(output_folder, image_file)

    imagen = tf.keras.preprocessing.image.load_img(input_image_path, target_size=(255,255))
    image_array = tf.keras.preprocessing.image.img_to_array(imagen)  
    # Load the image using tf.keras.preprocessing.image.load_img
    image_to_explain = tf.keras.preprocessing.image.load_img(input_image_path, target_size=(255,255))
    image_array = tf.keras.preprocessing.image.img_to_array(image_to_explain)



    explainer = lime_image.LimeImageExplainer(random_state=42)
    explanation = explainer.explain_instance(
        image_array,
        model.predict,
        hide_color=0,
        num_samples=1000
    )

    # Get the explanation image
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0],
        positive_only=True,
        num_features=3,
        hide_rest=True
    )
    image_with_mask = mark_boundaries(image_array / 255.0, mask)
    
 # Save the explained image to the output folder with higher DPI
    plt.figure(dpi=300)
    plt.imshow(image_with_mask)
    plt.axis('off')
    plt.savefig(output_image_path, bbox_inches='tight', pad_inches=0)
    plt.close()

    
```
